```python
import requests
import pandas as pd

# Fai la richiesta JSON al servizio REST
url = "http://dataservices.imf.org/REST/SDMX_JSON.svc/Dataflow"
response = requests.get(url)
data = response.json()

# Estrai la parte che contiene i dati di interesse
structure = data['Structure']
dataflows = structure['Dataflows']['Dataflow']

# Crea un elenco di dizionari per le righe del DataFrame
data_list = []

for dataflow in dataflows:
    row = {
        "SOURCE": dataflow.get('@agencyID'),
        "DATABASE": dataflow.get('@id'),
        "DATABASE_ID": dataflow['KeyFamilyRef'].get('KeyFamilyID'),
        "DATABASE DESCRIPTION": dataflow['Name']['#text'],
    }
    data_list.append(row)

# Crea il DataFrame
df = pd.DataFrame(data_list)

# Aggiungi la colonna "Last Refresh Time"
df['Last Refresh Time'] = pd.to_datetime('now')

# Visualizza il DataFrame
print(df)
```
